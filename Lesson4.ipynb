## Тема “Предел функции”

### Задание №1
Предложить пример функции, не имеющей предела в нуле и в бесконечностях.

$$f(x) = sin(x + \frac{1}{x})$$

```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.arange(-30, 30, 0.5)
y = np.sin(x + 1/x)

plt.axhline(linewidth=2.0, color="black")
plt.axvline(linewidth=2.0, color="black")

plt.plot(x,y)
plt.show()
```

### Задание №3
Исследовать функцию $f(x) = x^3 - x^2$ по плану:
1. Область задания и область значений.
2. Нули функции и их кратность.
3. Отрезки знакопостоянства.
4. Интервалы монотонности.
5. Четность функции.
6. Ограниченность.
7. Периодичность.

```python
x = np.linspace(-1, 2, 100)
y = x ** 2 * (x - 1)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.legend(labels=["$f(x) = x^3 - x^2$"])

plt.axhline(linewidth=2.0, color="black")
plt.axvline(linewidth=2.0, color="black")

plt.plot([0, 1], [0, 0], 'o')

plt.grid()
plt.show()
```

1. Область задания и область значений:
$D(f) = \mathbb{R}$

$E(f) = \mathbb{R}$
2. Нули функции: $f(x)=x^3 - x^2\implies f(x)=x^2(x - 1)\implies x_1 = 1, x_2 = 0$
3. Отрезки знакопостоянства: $(-\infty; 0), (0; 1);  (1; \infty)$
4. Интервалы монотонности: $ f(x) $ не убывает при $ x\in(-\infty,0]\cup[0.7,\infty) $ , убывает при $ x\in(0,0.7) $
5. Четность функции: не обладает свойством чётности
6. Ограниченность: не ограничена ни сверху, ни снизу
7. Периодичность: функция апериодична

### Задание №1
Найти предел:
a. 
$\lim _{x \rightarrow 0} \frac{\sin (2 x)}{4 x} = \frac{1}{2} \lim _{2x \rightarrow 0} \frac{\sin (2 x)}{2 x} = \frac{1}{2} \cdot 1 = \frac{1}{2}$

b.
$\lim _{x \rightarrow 0} \frac{x}{\sin x} = \frac{\lim _{x \rightarrow 0} 1}{\lim _{x \rightarrow 0} \frac{x}{\sin x}} = \frac{1}{1} = 1$

c.
$\lim _{x \rightarrow 0} \frac{x}{\arcsin x} = \lim _{t \rightarrow 0} \frac{\sin t}{t} = 1$

d.
$\lim _{x \rightarrow \infty}\left(1+\frac{k}{x}\right)^{x}=e^{k}$

$\lim _{x \rightarrow \infty}\left(\frac{4 x+3}{4 x-3}\right)^{6 x} = \lim _{x \rightarrow \infty}\left(1 + \frac{6}{4 x-3}\right)^{6 x} = \lim _{t \rightarrow \infty}\left(1 + \frac{6}{t}\right)^{\frac{6}{4}t+\frac{9}{2}} =\lim _{t \rightarrow \infty}\left(1 + \frac{6}{t}\right)^{\frac{6}{4}t} \cdot \lim _{t \rightarrow \infty}\left(1 + \frac{6}{t}\right)^{\frac{9}{2}} = e^{6\cdot\frac{6}{4}} \cdot 1 =  e^9$

e.
$\lim\limits_{x \to \infty}\frac{\sin{x}+\ln{x}}{x} = \lim\limits_{x \to \infty}\left(\frac{\sin{x}}{x} + \frac{\ln{x}}{x}\right) = \lim\limits_{x \to \infty} \frac{\sin{x}}{x} + \lim\limits_{x \to \infty}\frac{\ln{x}}{x}$

$\lim\limits_{x \to \infty} \frac{\sin{x}}{x} = \lim\limits_{x \to \infty} \left( \frac{-1}{x} \leq \frac{\sin{x}}{x} \leq \frac{1}{x} \right)$

$\lim\limits_{x \to \infty} \frac{-1}{x} = \lim\limits_{x \to \infty} \frac{1}{x} = 0 $

$\lim\limits_{x \to \infty}\frac{\ln{x}}{x} = ({\color{red}{\frac{\infty}{\infty}}}) = \lim\limits_{x \to \infty} \frac{\frac{d}{dx}[\ln{x}]}{\frac{d}{dx}[x]} = \lim\limits_{x \to \infty} \frac{\frac{1}{x}}{1} = \frac{\lim\limits_{x \to \infty}\frac{1}{x}}{\lim\limits_{x \to \infty} 1} = \frac{0}{1} = 0$

$\lim\limits_{x \to \infty} \frac{\sin{x}}{x} + \lim\limits_{x \to \infty}\frac{\ln{x}}{x} = 0 + 0 = 0 $
